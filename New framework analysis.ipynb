```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MaxNLocator
```

```python

# Define a custom formatter function for large numbers
def large_number_formatter(x, pos):
    """Converts large numbers using 'K' for thousands or 'M' for millions."""
    if x >= 1_000_000:
        return f'{int(x / 1_000_000)}M'
    elif x >= 1000:
        return f'{int(x / 1000)}K'
    return str(int(x))

def plotting_average_with_min_max(data, ylabel, xlabel, legendSet, title=None, saveAs=None):
    plt.figure(figsize=(6.2, 5.4))

    for d, l in zip(data, legendSet):
        avg = np.mean(d, axis=0)
        min_y = np.min(d, axis=0)
        max_y = np.max(d, axis=0)
        x = np.arange(len(avg))

        plt.plot(avg, label=l)
        plt.fill_between(x, min_y, max_y, alpha=0.2, label='_nolegend_')

    if title:
        plt.title(title)
    
    # Set x and y ticks font size
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)
    
    # Format x-axis ticks
    ax = plt.gca()
    ax.xaxis.set_major_formatter(FuncFormatter(large_number_formatter))
    ax.xaxis.set_major_locator(MaxNLocator(5))  # Control the maximum number of x-axis ticks to avoid clutter

    # Set labels and their font size
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    
    # Legend settings
    plt.legend(loc="lower right", fontsize=12)
    plt.tight_layout()
    
    # Save the figure if a filename is provided
    if saveAs:
        plt.savefig(saveAs, dpi=600)
    
    plt.show()
```
